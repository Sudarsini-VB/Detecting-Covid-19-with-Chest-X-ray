```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt

batch_size = 32

data_dir = "/kaggle/input/covid19-radiography-database/COVID-19_Radiography_Dataset/"

# Augmentation
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        validation_split=0.25,
        horizontal_flip=True)

# Training set
train = train_datagen.flow_from_directory(
    data_dir,
    target_size=(299, 299),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

# Validation set
val = train_datagen.flow_from_directory(
    data_dir,
    target_size=(299, 299),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

# Correct class names
class_names = ['COVID', 'Normal', 'Lung_Opacity', 'Viral Pneumonia']

# Function to plot images
def plotImages(images_arr, labels):
    fig, axes = plt.subplots(4, 4, figsize=(15,15))
    axes = axes.flatten()
    for img, lab, ax in zip(images_arr, labels, axes):
        ax.imshow(img)
        ax.set_title(class_names[np.argmax(lab)])
        ax.axis("off")
    plt.show()

# Get a batch of images
images, labels = next(train)

plotImages(images[:16], labels[:16])
```

```python
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.layers import Input, Dense, Dropout, Flatten
from tensorflow.keras.models import Sequential

# Load pretrained base model
base = InceptionV3(include_top=False, weights='imagenet', pooling='avg')

# Freeze base model
base.trainable = False

# Build model
model = Sequential()
model.add(Input(shape=(299, 299, 3)))
model.add(base)
model.add(Dropout(0.3))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(4, activation='softmax'))  # 4 classes

model.summary()
```
